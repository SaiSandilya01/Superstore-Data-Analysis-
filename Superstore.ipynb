```python
ss["PPU"]=ss["Sales"]/ss["Quantity"]
```

```python
ss["Order Date"]=pd.to_datetime(ss["Order Date"])
```

```python
ss.head()
```

```python
ss.info()
```

```python
ss["month"]=pd.DatetimeIndex(ss["Order Date"]).month
```

```python
ss["year"]=pd.DatetimeIndex(ss["Order Date"]).year
```

```python
ss.head()
```

```python
ss["year_month"]=ss["Order Date"].dt.to_period('M')
```

```python
ss.head()
```

```python
nss=ss.groupby("year_month")
nss.sum()["Sales"]
```

```python
cumm_sales=[]
for x in nss.sum()["Sales"]:
    cumm_sales.append(x)
cumm_sales    
```

```python
sorted(list(set(ss["year_month"])))
```

```python
rss=pd.DataFrame({"year_month":sorted(list(set(ss["year_month"]))),"Salespermonth":cumm_sales})
```

```python
rss
```

```python

rss["year_month"]=rss["year_month"].astype(str)
pd.to_datetime(rss.year_month)
rss.info()
```

```python
plt.plot_date(ss["Order Date"],ss["Sales"])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,5,5])
ax.bar(rss["year_month"],rss["Salespermonth"])
plt.show()
```
